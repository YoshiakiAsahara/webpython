```python
import pandas as pd
import numpy as np
```

リスト作成

```python
# リストからシリーズの作成
print('リストto Series')
s1 = pd.Series([0, 1, 2])
s2 = pd.Series(['ゼロ','イチ','二'])
s3 = pd.Series({0: 'boo', 1: 'foo', 2: 'woo'})
print(s1)
print(s2)
print(s3)
print()
d1=pd.DataFrame(np.arange(12).reshape(4,3),index=[12,13,14,15],columns=['ゼロ','イチ','二']))
print(d1)
print()
# 多次元のNumpy配列からのデータフレームの作成
d2 = pd.DataFrame(np.random.rand(12).reshape(4, 3), columns=['c1','c2','c3'])
print(d2)
print()
# 辞書からデータフレームの作成
d3 = pd.DataFrame({'Initial': ['B', 'F', 'W'], 'Name': ['boo', 'foo', 'woo']}, columns=['Name', 'Initial'])
print(d3)
```

データフレーム操作

```python
# Irisデータセットを読み込む
iris = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
print(iris.index)
print(len(iris.index))
print(iris.columns)
print(len(iris.columns))
print()
# 先頭5件のデータ
print(iris[:5])
# 末尾5件のデータ
print(iris[-5:])
# DataFrame の 'sepal_length' と 'species' の列の先頭5行のデータ
print(iris[['sepal_length', 'species']].head(5))
# DataFrameの2行目のデータ
print(iris.iloc[1])
# DataFrameの2行, 2列目のデータ
print(iris.iloc[1, 1])
# DataFrameの1から5行目と1から2列目のデータ
print(iris.iloc[0:5, 0:2])
# DataFrameの行インデックス5のデータ
print(iris.loc[5])
# DataFrameの行インデックス5と'sepal_length'列のデータ
print(iris.loc[5, 'sepal_length'])
# DataFrameの行インデックス1から5と'sepal_length'とspecies'の列のデータ
print(iris.loc[1:5, ['sepal_length','species']])
# DataFrameの行インデックス1から5と'sepal_length'とspecies'の列のデータ
print(iris.loc[1:5, ['sepal_length','species']])
# 'sepal_length' 列の値が7より大きく、'sepal_width'列の値が3より小さいデータ
print(iris[(iris['sepal_length'] > 7.0) & (iris['sepal_width'] < 3.0)])
# DataFrame に'mycolumn'という列を追加
iris['newcolumn'] = np.random.rand(len(iris))
iris.head(10)
```
